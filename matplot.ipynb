```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
sfood = pd.read_csv("C:/Users/moizr/Downloads/food_data.csv")

# Group by 'Calories', 'Fat', and 'Protein' and sum the values
grouped_data = sfood.groupby(['Calories', 'Fat', 'Protein']).sum().reset_index()

# Plot the stacked bars
plt.figure(figsize=(10, 6))
plt.stackplot(grouped_data.index, grouped_data['Calories'], grouped_data['Fat'], grouped_data['Protein'], labels=['Calories', 'Fat', 'Protein'])
plt.legend()
plt.xlabel('Group Index')
plt.ylabel('Amount')
plt.title('Nutritional Data Grouped by Calories, Fat, and Protein')
plt.show()
```
