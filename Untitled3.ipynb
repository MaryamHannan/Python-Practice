```python
# box plot 
import seaborn as sns
data2=sns.load_dataset("tips")
data2
sns.barplot(data=data2,x="day",y="tip",estimator='median')
plt.plot
```

```python
# hist plot 
import seaborn as sns
data2=sns.load_dataset("tips")
data2
sns.histplot(data=data2,x="day")
plt.plot
```

```python
# scatter plot 
sns.scatterplot(data=data2,x="total_bill" , y="tip")
```

```python

# Heatmap 
gp = data2.groupby("day", observed=False).agg({"tip": "mean"})
# gp=data2.groupby("day").agg({"tip": "mean"})
sns.heatmap(gp)
plt.show()
```

```python
# count plot seaborn 

sns.countplot(data=data2, x="day")
```

```python
# violin Plot 
with pd.option_context('mode.use_inf_as_na', True):
    sns.violinplot(data=data2, x="total_bill")
```

```python
# Pair plot 
with pd.option_context('mode.use_inf_as_na', True):
    sns.pairplot(data2,hue="day",diag_kind="hist")
    plt.show()
```

```python
# Strip plot 
sns.stripplot(data2,x="day",y="total_bill")
plt.show()
```

```python
# boxplot 
sns.boxplot(data2,x="tip")
plt.show()
```

```python

sns.boxplot(data2,y="tip",x="sex",orient="vertical")
plt.show()
```

```python
# CATPLOT
sns.catplot(data2,x="day",y='tip')
plt.show()
```

```python
# style & color in plot
sns.set_style(style='dark')
sns.barplot(data2,x="day",y="tip")
plt.show()
```

```python
# Multiple plot 

a=sns.FacetGrid(data2,col="smoker",height=2)
a.map(sns.barplot,'day','tip')
```

```python
# Relational plot 
sns.relplot(data=data2,x='tip',y='total_bill', hue='sex',col='day')
```

```python
# swarm plot
sns.swarmplot(data2,x='day',y='total_bill')
```

```python
 # kde plot 
sns.kdeplot(data=data2, x="total_bill")
```
